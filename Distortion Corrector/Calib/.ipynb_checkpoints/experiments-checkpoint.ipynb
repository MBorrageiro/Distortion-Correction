```python
from importlib import reload
import os
from datetime import datetime
import json
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib import collections  as mc

# For 3D visualisations
from PyQt5.QtWidgets import QApplication
import pyqtgraph.opengl as gl
import plotting

# My modules
import calib
```

Set up parameters

```python
board_shape = (8, 6)
board_edge_len = 0.045
camera_resolution = (2704, 1520)
int_dirs = [
    [
        "./tests/assets/intrinsic_calibration/images",
        "./tests/assets/intrinsic_calibration/data/points.json",
        "./tests/assets/intrinsic_calibration/data/camera.json"
    ]
]

```

## Intrinsic calibration 

```python
saved_points = []
saved_fnames = []

# Find calibration board corners in images
for i, [img_dir, output_file, _] in enumerate(int_dirs):
    print(f"Finding calibration board for images in {img_dir}")
    filepaths = sorted([os.path.join(img_dir, fname) for fname in os.listdir(img_dir) if fname.endswith(".jpg")])
    points, fpaths = calib.find_corners_images(filepaths, board_shape)
    fnames = [os.path.basename(f) for f in fpaths]
    saved_points.append(points.tolist())
    saved_fnames.append(fnames)
    for f in filepaths:
        if os.path.basename(f) not in fnames:
            print(f"Removing {f}")
            os.remove(f) #remove images where calibration board not found
    print(f"Saving points to {output_file}")
    created_timestamp = str(datetime.now())
    points = dict(zip(saved_fnames[i], saved_points[i]))
    data = {
        "created_timestamp": created_timestamp,
        "board_shape":board_shape,
        "images_dir": img_dir,
        "points": points
    }
    with open(output_file, "w") as f:
        json.dump(data, f)
```

```python
points = []
img_dirs = []
for [_, output_file, _] in int_dirs:
    with open(output_file, "r") as f:
        data = json.load(f)
        points.append(data["points"])
        img_dirs.append(data["images_dir"])
```

```python
for i in range(len(int_dirs)):
    corners = np.array(list(points[i].values()), dtype=np.float32)
    plt.figure(figsize=(16,9))
    
    for pts in corners:
        pts = pts.reshape(-1,2)
        cols = board_shape[0]
        rows = board_shape[1]
        edges = []
        for r in range(rows):
            for c in range(cols-1):
                edges.append(c+r*cols)
                edges.append(c+r*cols+1)
        for c in range(cols):
            for r in range(rows-1):
                edges.append(c+r*cols)
                edges.append(c+(r+1)*cols)
        lc = mc.LineCollection(pts[edges].reshape(-1,2,2), color='r',linewidths=0.5)

        plt.gca().add_collection(lc)
        plt.gca().set_xlim((0,camera_resolution[0]))
        plt.gca().set_ylim((camera_resolution[1], 0))
    plt.show()
```

### Fisheye camera calibration 

```python
created_timestamp = str(datetime.now())
for i, [_,_,camera_fpath] in enumerate(int_dirs):
    corners = np.array(list(points[i].values()), dtype=np.float32)
    obj_pts = calib.create_board_object_pts(board_shape, board_edge_len)
    k,d,r,t, used_corners = calib.calibrate_fisheye_camera(obj_pts, corners, camera_resolution)
    print("K:\n", k, "\nD:\n", d)#, "\nR:\n", r, "\nT:\n", t)
    data = {
        "created_timestamp": created_timestamp,
        "camera_type": "fisheye",
        "camera_resolution": camera_resolution,
        "k": k.tolist(),
        "d": d.tolist(),
    }
    with open(camera_fpath, "w") as f:
        json.dump(data, f)
    
    
    #Plotting
    plt.figure(figsize=(16,9))
    
    for pts in used_corners:
        pts = pts.reshape(-1,2)
        cols = board_shape[0]
        rows = board_shape[1]
        edges = []
        for row in range(rows):
            for col in range(cols-1):
                edges.append(col+row*cols)
                edges.append(col+row*cols+1)
        for col in range(cols):
            for row in range(rows-1):
                edges.append(col+row*cols)
                edges.append(col+(row+1)*cols)
        lc = mc.LineCollection(pts[edges].reshape(-1,2,2), color='r',linewidths=0.5)

        plt.gca().add_collection(lc)
        plt.gca().set_xlim((0,camera_resolution[0]))
        plt.gca().set_ylim((camera_resolution[1], 0))
    plt.show()
```

## Extrinsic calibration

Extract calibration board frames from a set of images for extrinsic calibration.

```python
board_shape = (9, 6)
board_edge_len = 0.088
ext_dirs = [
    [
        "./tests/assets/extrinsic_calibration/images/1",
        "./tests/assets/extrinsic_calibration/data/points_1.json"
    ],
    [
        "./tests/assets/extrinsic_calibration/images/2",
        "./tests/assets/extrinsic_calibration/data/points_2.json"
    ]
]
```

```python
calib = reload(calib)
```

```python
# Extract the points
saved_points = []
saved_fnames = []
count = {}

# Find calibration board corners in images
for [img_dir, output_file] in ext_dirs:
    print(f"Finding calibration board for images in {img_dir}")
    filepaths = sorted([os.path.join(img_dir, fname) for fname in os.listdir(img_dir) if fname.endswith(".jpg")])
    points, fpaths = calib.find_corners_images(filepaths, board_shape, window_size=5) #smaller window size due to being further away from camera
    fnames = [os.path.basename(f) for f in fpaths]
    saved_points.append(points.tolist())
    saved_fnames.append(fnames)
    for fname in fnames:
        if fname in count:
            count[fname] += 1
        else:
            count[fname] = 1
    for f in filepaths:
        if os.path.basename(f) not in fnames:
            print(f"Removing {f}")
            os.remove(f) #remove images where calibration board not found
    print()
            
# Remove images which don't have corresponding images from other cameras
created_timestamp = str(datetime.now())
invalid_names = [fname for fname, value in count.items() if value <= 1]
for i, [img_dir, output_file] in enumerate(ext_dirs):
    for j, fname in enumerate(saved_fnames[i]):
        if fname in invalid_names:
            print(f"Removing {fname} from {img_dir}")
            os.remove(os.path.join(img_dir,fname)) #remove images that have no correspondences to other cameras
            saved_fnames[i].pop(j)
            saved_points[i].pop(j)
    points = dict(zip(saved_fnames[i], saved_points[i]))
    data = {
        "created_timestamp": created_timestamp,
        "board_shape":board_shape,
        "images_dir": img_dir,
        "points": points
    }
    with open(output_file, "w") as f:
        json.dump(data, f)
```

```python
def cam1_pts_to_cam2(r, t, pts_3d_cam1):
    pts = pts_3d_cam1.reshape((-1,3,1))
    pts_3d_cam2 = (r@pts + t).reshape((-1,3))
    return pts_3d_cam2
```

```python
class Scene:
    
    def __init__(self):
        self.app = QApplication.instance()
        if self.app == None:
            self.app = QApplication([])
        self.view = gl.GLViewWidget()
        self.view.setBackgroundColor("#FFFFFF")
        self.view.show()
    
    
    def rodrigues_to_vec(self, r):
        ang = 180/np.pi*np.linalg.norm(r)
        return (ang, *r)
        

    def plot_camera(self, r, t):
        r = cv2.Rodrigues(r)[0]
        r = self.rodrigues_to_vec(r)
        cam = plotting.create_camera()
        cam.rotate(*r)
        cam.translate(*t)
        self.view.addItem(cam)
        
        
    def plot_calib_board(self, r, t, board_shape, board_edge_len):
        obj_pts = calib.create_board_object_pts(board_shape, board_edge_len)
        calib_board = plotting.create_grid(obj_pts, board_shape)
        calib_board.translate(*t)
        r = rodrigues_to_vec(r)
        calib_board.rotate(*r)
        self.view.addItem(calib_board)
        

        c1 = plotting.create_camera()
        c1.rotate(-90, 1,0,0)
        view.addItem(c1)

        rot = cv2.Rodrigues(r)[0]
        print(rot)
        ang = 180/np.pi*np.linalg.norm(rot)
        c2 = plotting.create_camera()
        c2.rotate(-90, 1,0,0)
        c2.rotate(-ang, *rot, local=True)
        c2.translate(*-t)
        view.addItem(c2)

    def run(self):
        self.app.exec_()
```

```python
#camera r and t

#cam 1: rotate -90 degrees around x axis, positioned at origin
RI_1 = np.array([[1, 0, 0],
                 [0, 0, -1],
                 [0, 1, 0]], dtype=np.float32)
R1_I = RI_1.T
TI_1 = np.array([[0, 0, 0]], dtype=np.float32).T
T1_I = -TI_1

#cam 2: rotate -90 degrees around x axis, positioned at origin
RI_2 = r@RI_1
R2_I = RI_2.T
TI_in_2 = r@TI_1 + t
T2_in_I = R2_I@TI_in_2

print(T1_I)
print(T2_I)

scene = Scene()
scene.plot_camera(R1_I, T1_I)
scene.plot_camera(R2_I, TI_in_2)

scene.run()
```
